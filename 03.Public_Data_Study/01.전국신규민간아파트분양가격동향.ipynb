#### < 규모구분을 전용면적 컬럼으로 변경 >
- 데이터 조작
    - 데이터에서 '전용면적' 제거
    - ex) 전용면적 60㎡초과 85㎡이하 → 60㎡~85㎡

=> 데이터 용량 줄여서 메모리를 효율적으로 사용하기 위해 

```python
df_last['규모구분'].unique()   # 컬럼의 고유 데이터 값 
```

```python
# 데이터 조작 - 전용면적 60㎡초과 85㎡이하 → 60㎡~85㎡ 와 같은 형태로
df_last['전용면적'] = df_last['규모구분'].str.replace('전용면적', '').str.replace('초과', '~').str.replace('이하', '')
df_last['전용면적']
```

```python
# 중간 공백 제거 # 앞뒤 공백 제거
df_last['전용면적'] = df_last['전용면적'].str.replace(" ", "").str.strip()
df_last['전용면적']
```

### ▶ group by로 데이터 집계하기
- groupby를 통해 데이터를 그룹화해서 연산

```python
# 지역명으로 컬럼별 평균값 구하기
df_last.groupby(['지역명']).mean()
```

```python
# 지역명으로 평당분양가격의 평균 구하기
# max, min, describe 사용 가능
df_last.groupby(['지역명'])['평당분양가격'].mean()
```

```python
# 전용면적으로 평당분양가격의 평균 구하기
df_last.groupby(['전용면적'])['평당분양가격'].mean()
```

```python
# 지역명, 전용면적으로 평당분양가격의 평균 구하기
# groupby()에 입력하는 컬럼명 순으로 그룹화 함
df_last.groupby(['지역명','전용면적'])['평당분양가격'].mean()
```

```python
# unstack() : groupby()에 입력한 마지막 인자가 컬럼값으로 오는 데이터프레임 반환
# round() : 반올림
df_last.groupby(['지역명','전용면적'])['평당분양가격'].mean().unstack().round()
```

```python
# 연도, 지역명으로 평당분양가격의 평균 구하기
# .T : 행과 열 자리 바꿈
g = df_last.groupby(['연도','지역명'])['평당분양가격'].mean()
g
# g.unstack().T
```

### ▶ pivot table 로 데이터 집계하기

```python
# pivot_table(데이터, 인덱스, 컬럼, 집계함수)
pd.pivot_table(df_last, index=['지역명'], values=['평당분양가격'], aggfunc='mean')
```

```python
# pd.pivot은 데이터의 형태만 바꿈(연산X) - aggfunc 없음
pd.pivot_table(df_last, index='전용면적', values='평당분양가격')
```

```python
# df_last.groupby(['지역명','전용면적'])['평당분양가격'].mean()
df_last.pivot_table(index='지역명', columns='전용면적', values='평당분양가격')
```
