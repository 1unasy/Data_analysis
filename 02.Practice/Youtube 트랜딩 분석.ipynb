# 인기 영상 키워드 분석

````python
cols = ["title", "view_count"]
df_title = df_yt[cols]
df_title

# 정렬
df_sorted = df_title.sort_values(by="view_count", ascending=False)
df_sorted.head()
````

````python
# 제목 중복 제거
df_sorted = df_sorted.drop_duplicates(['title'], keep='first')
df_sorted.head()
````

# 제목만 가져오기

````python
df_hot = df_sorted.copy()
df_hot['title'].values
````

````python
df_title_kor = df_hot['title'].str.replace("[^가-힣\s]", "")
df_hot['title_refined'] = df_title_kor
df_hot
````

- ^ = 그게 아닌 것 like `not`
- 가-힣 = 완성형 한글 모드 (완성형 한글 처음부터 끝까지, 가나다라 ... 힣)
- \s = 공백 문자 1개 (spacebar, enter 키 한 칸)
- ^가-힣\s = 완성형 한글 전체나 공백이 아닌 것을 선택하겠다
- 영어는 a-zA-Z\s 치면 됌
- str.replace("[^가-힣\s]", "") : 완성형 한글 전체나 공백이 아닌 것을 선택해서 " "(비어있는 문자열로)으로 치환하겠다.
- 즉, replace " "은 삭제를 의미한다

````python
import re

df_hot = df_hot[ df_hot['title_refined'].apply(lambda x : re.sub('[^가-힣]', '', x)) != '' ]
df_hot
````

# 제목에서 단어 추출, 빈도 수 구하기
- 인기 영상 키워드 분석하기

````python
from soynlp.word import WordExtractor
````

- 단어 점수 : 단어가 될 수 있는 수치
```
예시 1. 나는 사과를 먹는다. -> 명사 추출 -> [나, 사과]
예시 2. 아이오아이는 가수다. -> 명사 추출 -> [가수]
```

soynp는 연속되어지는 문자의 빈도, 경계점수를 활용해서 신조어를 단어로 인식한다.
- 연속되어지는 문자의 빈도 : 응집도 ( cohension )
- 경계점수 ( boundary entropy )

- 응집도

뉴스에서 아이오아이가 처음 등장했다.

.[속보].아이오아이의 앨범 ~~

.[2보].아이오아이가 콘서트 개최 ~~

-> 아이오아이 라는 특정 단어가 자주 등장 = 응집도가 높다 = 단어로 인식

- 경계점수

아이오아이 가

아이오아이 는

아이오아이 의

아이오아이 를

아이오아이 에게

아이오아이 에서

entropy가 높다 = 혼잡하다, 복잡하다

-> 안정적인 문자열 + 불안정한 character가 붙으면 그 앞에 있는 것들을 단어라고 봄

````python
word_ext = WordExtractor(
    min_frequency = 10, # 최소 등장 빈도
    min_cohesion_forward=0.05,# 앞쪽에 등장하는 단어의 최소 cohension 스코어
    min_right_branching_entropy = 0.0 # 특정 단어 뒷쪽에 등장하는 문자에 대한 혼잡도의 최소값
)

word_ext.train(df_hot['title_refined'].values) # 문장들을 입력시켜서 단어 훈련
words = word_ext.extract() # 단어 추출
words
````

- cohesion_forward = 0 ; 앞글자가 없으니까 0

````python
for word in [words['방'], words['방탄'], words['방탄소'], words['방탄소년'], words['방탄소년단']]:
    print(word.cohesion_forward, "\t\t\t", word.right_branching_entropy)
````

- cohesion : 
- entropy : 검색 단어 뒤에 올 조사에 얼마나 다양한 변수가 존재하는지

````python
from soynlp.tokenizer import LTokenizer # 왼편에 있는 ~ 만 남기겠다

# 응집도 점수만 뽑기
cohesion_score = { word : score.cohesion_forward for word, score in words.items() }

# 응집도 점수를 기반으로 해서 Tokenizer 만들기
tokenizer = LTokenizer(scores=cohesion_score)
````

````python
df_hot['tokenize'] = df_hot['title_refined'].apply(lambda x : tokenizer.tokenize(x, remove_r=True))
df_hot
# tokenize() 함수 한글을 추출하겠다는 뜻
# remove_r = 조사를 제거
````

# 50만 조회수 이상만 추출

````python
hot_mask = df_hot['view_count'] >= 500000
df_hot_top = df_hot[hot_mask]
df_hot_top
````

````python
df_hot_top['tokenize'] # 리스트 형태로 들어있는 배열 -> 배열 안쪽 원소들을 하나로 병합시켜보자 !! = 파이썬 기술 중 하나
````

````python
# [] 비어있는 배열에 tokenize 컬럼에 있는 원소들을 다 더해줌
words = sum(df_hot_top['tokenize'], [])
words 
````

# 단어들의 빈도를 파악

````python
from collections import Counter

count = Counter(words) # 배열 안쪽 원소들이 등장한 빈도
words_dict = dict(count)

print(words_dict)
````
